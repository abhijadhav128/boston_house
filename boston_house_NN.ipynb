```python
Y=A[["SalePrice"]]
X=A.drop(columns=["Id","SalePrice"],axis=1)
```

```python
cat = []
con = []
for i in X.columns:
    if(X[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Ynew=pd.DataFrame(le.fit_transform(Y.SalePrice),columns=["Loan_Status"])
Ynew.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
X1 = pd.DataFrame(mm.fit_transform(X[con]),columns=con)
X2 = pd.get_dummies(X[cat])
Xnew = X1.join(X2)
```

```python
Xnew.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
nn=Sequential()
nn.add(Dense(units=100,input_dim=(len(Xnew.columns))))
nn.add(Dense(100,activation="relu"))
nn.add(Dense(1,activation="relu"))
```

```python
nn.compile(loss="mse")
model=nn.fit(Xnew,Ynew,validation_split=0.2,epochs=10)
```

```python
tr_err = model.history['loss']
ts_err = model.history['val_loss']
```
